```python
import pandas as pd
df = pd.read_csv('Data/clean.csv')
```

```python
print(df.head())
```

```python
from datasets import Dataset

dataset = Dataset.from_pandas(df)
print(dataset)
```

```python
import spacy
nlp = spacy.load('fr_core_news_sm')
# Boucler sur chaque texte dans le dataset
for text in dataset['text']:
    doc = nlp(text)
    for token in doc:
        print(token.text, token.lemma_, token.pos_, token.dep_)
```

```python
from prettytable import PrettyTable

table = PrettyTable(["Token", "Lemma", "POS", "Dep"])

for text in dataset['text']:
    doc = nlp(text)
    for token in doc:
        table.add_row([token.text, token.lemma_, token.pos_, token.dep_])

print(table)
```
